```python
import pandas as pd
```

```python
df = pd.read_csv('../../Cleaned Datasets/NSL-KDD/Cleaned_NSLKDD.csv')
```

```python
df
```

```python
features_to_be_normalised = []
feature_list = df.columns
for col in feature_list:
    if (df[col].dtype == 'int64' or df[col].dtype == 'float64') and (df[col].max() - df[col].min() > 1000):
        features_to_be_normalised.append(col)
print(features_to_be_normalised)
```

```python
test_df = df[features_to_be_normalised]
normalized_df= (test_df-test_df.min())/(test_df.max()-test_df.min())
```
